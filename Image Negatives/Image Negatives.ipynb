# Intensity Transformations--Image Negative

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Assuming our photo is 8 bit, the density values are between 0-255. If we want to get the negative of a photograph,
#it is necessary to equalize it to the value that will complete it to the maximum intensity level.
#mathematical expression   ==>     s = L-1-r  (input value)



```python
image= cv2.imread("sunflower.jpg", 0)
plt.imshow(image, cmap="gray")
```

Negative Photo Function
    

```python
def image_negative(image):
    L= np.max(image)
    negative_p= L-image
    return negative_p
```

```python
negative_photo = image_negative(image)
plt.imshow(negative_photo ,cmap="gray")
plt.show()
```

Let's see the photos side by side.

```python

side= np.hstack((image, negative_photo))
plt.imshow(side, cmap= "gray")
plt.show()
```
