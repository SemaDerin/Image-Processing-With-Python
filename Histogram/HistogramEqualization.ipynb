# Histogram Equalization

Hello everyone! We will look at how to equalize Histograms using a function available in the OpenCv library to equalize histograms.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cv2_main():
    image =cv2.imread("sunflower.jpg",0)
    hist_eq_image= cv2.equalizeHist(image)
```

```python
side_by_side= np.hstack((image,hist_eq_image))
plt.imshow(side_by_side, cmap="gray")
plt.show()
```
