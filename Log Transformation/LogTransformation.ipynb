# Log Transformation


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

The log transform distributes low intensity values in the narrow range over a wider range. 
It also compresses high intensity values into a narrower range.


```python
image= cv2.imread("log.jpg", 0)
```

unsigned int (255+1)= 0 ------> astype(float)

```python
def log_transformation(r,c):
    r= r.astype(float)
    s= c*np.log(1+r)
    return s.astype(np.uint8)
```

Stack function( for see side by side)


```python
def stack(*args):
    return np.hstack(args)
```

to express density values between 0-255

```python
def normalization(image):
    s= image.astype(float)
    s-=np.min(s)
    s/=np.max(s)
    return (s*255).astype(np.uint8)
```

If we apply it for our function photos

```python
log_photo= log_transformation(image,c=1)
log_photo= normalization(log_photo)
```

let's see them side by side and interpret the differences

```python
side_by_log= stack(image, log_photo)
plt.imshow(side_by_log, cmap="gray")
plt.show()
```

As can be seen in the picture, spots with low intensity values were scattered over an even wider range.
