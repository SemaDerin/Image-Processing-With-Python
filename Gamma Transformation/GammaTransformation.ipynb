# Gamma Transformation

Let's start by importing the libraries we will need first.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Let's transfer the image we will use to the Python environment.

```python
image = cv2.imread("image.jpg", 0 )
```

Gamma Transformation Function:

A small function we wrote to see the images side by side

```python
def stack(*args):
    return np.hstack(args)
```

Scaling function to keep intensity values between 0-255 after applying Gamma Transform.

```python
def normalization(image):
    s= image.astype(float)
    s-=np.min(s)
    s/=np.max(s)
    return (s*255).astype(np.uint8)
```

Gamma Transformation Function :

```python
def gamma_transformation(r,c,gamma):
    r= r.astype(float)
    s= c*r**gamma
    s= normalization(s)
    return s
```

We choose c=1 and gamma values are 3.0 - 4.0 and 5.0

```python
c=1
gamma_values= [3.0,4.0,5.0]
gamma_photos= []
for gamma in gamma_values:
    gamma_photo= gamma_transformation(image, c=c, gamma=gamma)
    gamma_photos.append(gamma_photo)
    


line1= stack(image, gamma_photos[0])
line2= stack(*gamma_photos[1:])
grid = np.vstack((line1,line2))


plt.imshow(grid, cmap="gray")
plt.show()
```
